# Python function to automate all the previous steps (insertion of the form def and return in the formulation)

```python
def financial_analysis(ticker):

    ''' Data 2020/2024 '''

    # Retrieving company data from Yahoo Finance
    data = yf.Ticker(ticker)

    # --- 1. Displaying general information ---
    print(f"\n General information about {ticker}:")
    print(data.info)

    # --- 2. Retrieval and formatting of financial data ---
    # Balance sheet
    balance_sheet = data.balance_sheet.T
    balance_sheet.index = pd.to_datetime(balance_sheet.index).year

    # Income statement
    income_statement = data.financials.T
    income_statement.index = pd.to_datetime(income_statement.index).year

    # Cash flow statement
    cash_flow = data.cashflow.T
    cash_flow.index = pd.to_datetime(cash_flow.index).year

    # Displaying financial data
    print(f"\n Balance Sheet of {ticker}:\n", balance_sheet.head())
    print(f"\n Income Statement of {ticker}:\n", income_statement.head())
    print(f"\n Cash Flow Statement of {ticker}:\n", cash_flow.head())

    # --- 3. Calculation of key financial ratios ---
    # Checking available columns
    print("\n Available columns in the balance sheet:")
    print(balance_sheet.columns)

    # Creating the financial ratios DataFrame
    financial_data = pd.DataFrame(index=income_statement.index)

    # Checking and correcting column names
    equity_col = "Stockholders Equity" if "Stockholders Equity" in balance_sheet.columns else "Total Stockholder Equity"
    assets_col = "Total Assets" if "Total Assets" in balance_sheet.columns else "Total Assets"
    liabilities_col = "Total Liabilities Net Minority Interest" if "Total Liabilities Net Minority Interest" in balance_sheet.columns else "Total Liabilities"
    debt_col = "Total Debt" if "Total Debt" in balance_sheet.columns else "Long Term Debt"

    # Calculating financial ratios
    financial_data["ROE"] = income_statement["Net Income"] / balance_sheet[equity_col]
    financial_data["ROA"] = income_statement["Net Income"] / balance_sheet[assets_col]
    financial_data["Net Margin"] = income_statement["Net Income"] / income_statement["Total Revenue"]
    financial_data["Liquidity Ratio"] = balance_sheet[assets_col] / balance_sheet[liabilities_col]
    financial_data["Debt Ratio"] = balance_sheet[debt_col] / balance_sheet[assets_col]

    # Displaying calculated ratios
    print(f"\n Financial Ratios of {ticker}:\n", financial_data)

    # --- 4. Visualization of financial ratios ---
    sns.set_style("darkgrid")

    # Creating a multi-plot figure
    fig, axes = plt.subplots(5, 1, figsize=(14, 10))

    # ROE
    sns.lineplot(x=financial_data.index, y=financial_data["ROE"], marker="o", ax=axes[0], color="blue")
    axes[0].set_title("ROE (Return on Equity) Evolution")

    # ROA
    sns.lineplot(x=financial_data.index, y=financial_data["ROA"], marker="o", ax=axes[1], color="red")
    axes[1].set_title("ROA (Return on Assets) Evolution")

    # Net Margin
    sns.lineplot(x=financial_data.index, y=financial_data["Net Margin"], marker="o", ax=axes[2], color="green")
    axes[2].set_title("Net Margin Evolution")

    # Liquidity Ratio
    sns.lineplot(x=financial_data.index, y=financial_data["Liquidity Ratio"], marker="o", ax=axes[3], color="purple")
    axes[3].set_title("Liquidity Ratio Evolution")

    # Debt Ratio
    sns.lineplot(x=financial_data.index, y=financial_data["Debt Ratio"], marker="o", ax=axes[4], color="orange")
    axes[4].set_title("Debt Ratio Evolution")

    # Adjusting the display
    plt.tight_layout()
    plt.show()

    # --- 5. Stock price evolution ---
    data_stock = data.history(period="5y")

    plt.figure(figsize=(10,5))
    plt.plot(data_stock.index, data_stock["Close"], label=f"{ticker} Closing Price", color="blue")
    plt.xlabel("Date")
    plt.ylabel("Price ($)")
    plt.title(f"Stock Price Evolution of {ticker}")
    plt.legend()
    plt.grid()
    plt.show()

    # Displaying the first 5 rows of stock prices
    print(f"\n Stock Prices of {ticker}:")
    print(data_stock.head())

    return (financial_data["ROE"], financial_data["ROA"], financial_data["Net Margin"],
            financial_data["Liquidity Ratio"], financial_data["Debt Ratio"])
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns

def compare_ratios(tickers):
    """
    Compare the key financial ratios of two companies.

    Parameters:
    - tickers (list_of_str): Stock symbols of the companies.

    Returns:
    - A DataFrame containing the financial ratios of both companies.
    - A comparative chart of the ratios.
    """

    companies = {company: yf.Ticker(company) for company in tickers}
    ratios = {}

    for ticker, company in companies.items():
        # Retrieving financial data
        balance_sheet = company.balance_sheet.T
        income_statement = company.financials.T

        # Checking columns to avoid errors
        equity_col = "Stockholders Equity" if "Stockholders Equity" in balance_sheet.columns else "Total Stockholder Equity"
        assets_col = "Total Assets" if "Total Assets" in balance_sheet.columns else "Total Assets"
        liabilities_col = "Total Liabilities Net Minority Interest" if "Total Liabilities Net Minority Interest" in balance_sheet.columns else "Total Liabilities"
        debt_col = "Total Debt" if "Total Debt" in balance_sheet.columns else "Long Term Debt"

        # Calculating financial ratios
        ratios[ticker] = {
            "ROE": income_statement["Net Income"].iloc[0] / balance_sheet[equity_col].iloc[0],
            "ROA": income_statement["Net Income"].iloc[0] / balance_sheet[assets_col].iloc[0],
            "Net Margin": income_statement["Net Income"].iloc[0] / income_statement["Total Revenue"].iloc[0],
            "Liquidity Ratio": balance_sheet[assets_col].iloc[0] / balance_sheet[liabilities_col].iloc[0],
            "Debt Ratio": balance_sheet[debt_col].iloc[0] / balance_sheet[assets_col].iloc[0]
        }

    # Converting to DataFrame
    df_ratios = pd.DataFrame(ratios)

    # Displaying the comparative table
    print("\nComparison of Financial Ratios:\n")
    print(df_ratios)

    # Visualizing the ratios in a chart
    sns.set_style("darkgrid")
    df_ratios.plot(kind="bar", figsize=(10,5))
    plt.title("Comparison of Financial Ratios")
    plt.xlabel("Company")
    plt.ylabel("Ratio Value")
    plt.legend(title="Ratios", bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.xticks(rotation=0)
    plt.grid()
    plt.show()

    return df_ratios
```
